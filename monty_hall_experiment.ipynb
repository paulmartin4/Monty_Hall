```python
%matplotlib inline

import matplotlib.pyplot as plt
from Monty_Hall_Fonction import Monty_Hall

print("Voici les resultats de l'experience de Monty Hall \npour 10 000 parties jouées pour chaque stratégie.\n")
plot = plt.bar([1, 2], Monty_Hall(), 
        tick_label = ["Changer","Garder"])
```

```python
gains_changer = []
gains_garder = []
samples = [1000, 10000, 20000, 50000, 80000, 100000]
for tours in samples:
    gains = Monty_Hall(tours)
    gains_changer.append(gains[0])
    gains_garder.append(gains[1])
    
print(f"Et ici ceux pour respectivement {samples} parties.")
figure = plt.figure()
plot = plt.scatter(samples, gains_changer)
plot = plt.scatter(samples, gains_garder)
```
